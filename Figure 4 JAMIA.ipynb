# Data

```python
cols_of_interest = ['input file name', 'correct diagnosis', 'gpt diagnosis', 'Score (0-5)', 'Rank', 'Score (Peter)', 'Score (Justin)', 'Range']

fpath_narrative = 'data/JAMARERUN on 2023-08-25 ChatGPT4 Diagnosis - text without discussion - txt_cases_results.tsv'
narrative_aug_2023 = pd.read_csv(fpath_narrative, sep='\t', usecols=cols_of_interest)
narrative_aug_2023['Src'] = 'aug2023'
narrative_aug_2023
```

```python
fpath_narrative = 'data/JAMARUN on 2023-06 text without discussion - txt_cases_results.tsv'
narrative_jul_2023 = pd.read_csv(fpath_narrative, sep='\t', usecols=cols_of_interest)
narrative_jul_2023['Src'] = 'jul2023'
narrative_jul_2023
```

```python
cols = ['input file name', 'Src', 'Score (0-5)', 'Rank']
df = pd.concat([narrative_aug_2023[cols], narrative_jul_2023[cols]]).sort_index().rename(columns={'Score (0-5)':'Score'})
df
```

Skip the cases where the score is N/A

```python
print(df.loc[df.Score.isna()])
df = df.loc[~df.Score.isna()]
```

Convert to `str`s

```python
df = df.astype({'Score': int, 'Rank': int})
df
```

```python
fig, ax = plt.subplots(figsize=(6,4.5), dpi=120)

data = df.astype({'Score': str}).groupby(['Src', 'Score']).size()
data.name='Count'
data = data.reset_index()

sns.barplot(data=data, x='Score', y='Count', hue='Src', 
            palette=['#3465a4', '#110942'],
            hue_order=['jul2023', 'aug2023'],
            ax=ax)
ax.grid(axis='y')
ax.set(ylim=(0, 70))
ax.legend_.set(title=None)
ax.legend_.get_texts()[0].set(text='NARR - July 2023')
ax.legend_.get_texts()[1].set(text='NARR - August 2023')
fig.tight_layout()
fig.savefig('Fig4.png')
```
